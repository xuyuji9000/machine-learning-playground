```python
import numpy as np
```

```python
def cross_entropy(predictions, targets):
    # Clip predictions to avoid log(0) which results in infinity
    predictions = np.clip(predictions, 1e-12, 1.0)
    return -np.sum(targets * np.log(predictions)) / predictions.shape[0]
```

```python

# Example: 3 classes (Cat, Dog, Bird)
# True label is 'Dog' (index 1)
true_labels = np.array([0, 1, 0])

# Case A: Model is confident and correct
pred_a = np.array([0.1, 0.8, 0.1])

# Case B: Model is confident but WRONG
pred_b = np.array([0.7, 0.1, 0.2])

print(f"Loss A (Correct): {cross_entropy(pred_a, true_labels):.4f}")
print(f"Loss B (Wrong):   {cross_entropy(pred_b, true_labels):.4f}")
```
