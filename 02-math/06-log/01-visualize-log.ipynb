```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot the log range `(0.01, 1]`

```python


x = np.linspace(0.01, 1, 500)
y = - np.log(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = - log(x)")
# plt.axhline(0, color='red', linestyle='--', linewidth=2, label='y = 0')  # Emphasize the y=0 line
plt.title("Natural Logarithm Function")
plt.xlabel("x")
plt.ylabel("- log(x)")
plt.grid(True)
plt.legend()
plt.show()
```
