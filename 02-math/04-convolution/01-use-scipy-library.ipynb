```python
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
```

```python
# 1. Create a simple "image" (10x10) 
# A dark left side (0s) and a light right side (1s)
image = np.zeros((10, 10))
image[:, 5:] = 1

print(image)

plt.imshow(image, cmap='gray')
plt.title('Simple Image')
plt.axis('off')
plt.show()

```

```python
# 2. Define a Vertical Edge Detection Kernel (Sobel)
# This kernel looks for differences between left and right pixels
kernel = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
```

```python
# 3. Perform the convolution
# 'valid' means we only calculate where the kernel fits completely
output = convolve2d(image, kernel, mode='valid')

output
```

```python
# Displaying the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, cmap='gray')
ax1.set_title("Original Image (Edge in middle)")
ax2.imshow(output, cmap='gray')
ax2.set_title("Convolution Result (Edge Highlighted)")
plt.show()
```
