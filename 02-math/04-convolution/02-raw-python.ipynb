```python
import numpy as np
```

```python

def simple_convolution(image, kernel):
    # Get dimensions
    img_h, img_w = image.shape
    ker_h, ker_w = kernel.shape
    
    # Calculate output dimensions
    out_h = img_h - ker_h + 1
    out_w = img_w - ker_w + 1
    output = np.zeros((out_h, out_w))

    # Slide the kernel over the image
    for i in range(out_h):
        for j in range(out_w):
            # Extract the current patch (the "window")
            window = image[i:i+ker_h, j:j+ker_w]
            # Multiply and sum
            output[i, j] = np.sum(window * kernel)
            
    return output
```

```python


# --- Example: Vertical Edge Detection ---
# A "dummy" image with a clear vertical transition
image = np.array([
    [10, 10, 10, 0, 0, 0],
    [10, 10, 10, 0, 0, 0],
    [10, 10, 10, 0, 0, 0],
    [10, 10, 10, 0, 0, 0]
])

import matplotlib.pyplot as plt

plt.imshow(image, cmap='gray', interpolation='none')
plt.title("Input Image")
plt.show()

```

```python

# Sobel filter for detecting vertical edges
kernel = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])

result = simple_convolution(image, kernel)

print("Detected Edges:\n", result)

plt.imshow(result, cmap='gray', interpolation='none')
plt.title("Detected Edges")
plt.colorbar()
plt.show()
```
