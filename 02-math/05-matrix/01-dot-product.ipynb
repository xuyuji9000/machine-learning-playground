```python
import numpy as np

```

```python

# Define two vectors (1D arrays) as lists
vector1 = [1, 2, 3]
vector2 = [4, 5, 6]

# Compute the dot product
# The formula is (1*4) + (2*5) + (3*6) = 4 + 10 + 18 = 32
result = np.dot(vector1, vector2)

# Print the result
print(f"The dot product is: {result}")
```
